```python
import numpy as np
import pandas as pd
```

```python
class K_mean:
  def __init__(self,K,maxiter):
    self.K = K
    # Định nghĩa số vòng lặp tối đa cho phép
    self.maxiter = maxiter
  def fit(self,X):
    self.X = X
    self.new_centroids = self.X[np.random.choice(self.X.shape[0],self.K,replace=False)]
    matrix_distance = np.stack((np.zeros((self.X.shape[0],1)),)*self.K,axis = 2)
    print(self.new_centroids)
    print('-------')
    self.old_centroids = np.zeros(self.new_centroids.shape)
    while not np.allclose(self.new_centroids,self.old_centroids) and self.maxiter > 0:
      for i in range(self.K):
        matrix_distance[:,:,i] = np.sqrt(((self.new_centroids[i].reshape(1,-1) - self.X)**2).sum(axis = 1)).reshape(self.X.shape[0],-1)
      matrix_clusterings = np.argmin(matrix_distance,axis=2)
      # Chỗ này không được gán trực tiếp hai mảng bằng nhau => mảng được gán sẽ trỏ đến vùng nhớ với mảng gán
      self.old_centroids = self.new_centroids.copy()
      for i in range(self.K):
        self.new_centroids[i] = np.mean(self.X[matrix_clusterings ==i],axis=0)
      print(f"old_centroids = {self.old_centroids}")
      print("--------")
      print(f"new_centroids = {self.new_centroids}")
      print('--------')
      print(not np.allclose(self.new_centroids,self.old_centroids))
      self.maxiter-=1
      self.Label = matrix_clusterings
```

```python
K_mean_clustering = K_mean(K=2,maxiter=100)
```

```python
X = np.array([1.4,1,1.5,3.1,3.8,4.1]).reshape(-1,1)
```

```python
K_mean_clustering.fit(X)
```
