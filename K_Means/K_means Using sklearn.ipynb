```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
FILE_PATH = '/content/drive/MyDrive/AIO 2024/Module_3/Week 2/T6/Data/iris_2D.csv'
```

```python
# Đọc dữ liệu từ file csv
df = pd.read_csv(FILE_PATH)
df
```

```python
data =df.drop(columns=['Label'])
data
```

```python
data = data.to_numpy()
data
```

```python
# khởi tạo tham số cần thiết
K = 2 # Số cụm cần phân chia dữ liệu
maxiter = 100 # Số vòng lặp tối đa để phân chia cụm
```

```python
# khởi tạo mô hình
K_means = KMeans(n_clusters=K,max_iter=maxiter)
```

```python
# Huấn luyện mô hình
K_means.fit(data)
```

```python
# phân cụm dữ liệu
labels = K_means.labels_
labels
```
